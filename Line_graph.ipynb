```python
import numpy as np

# Function to plot monthly average and yearly average
def plot_with_monthly_and_yearly_average(df, location, color, contrast_color):
    plt.figure(figsize=(16, 6))

    # Calculate a unique identifier for each year and month
    df['year_month'] = df['year'] + (df['month'] - 1) / 12  # Create a new column for year and month

    # Group by year and month to calculate the average sea level for each month
    df_monthly_avg = df.groupby('year_month')['sea_level'].mean().reset_index()

    # Group by year to calculate yearly averages
    df_yearly_avg = df.groupby('year')['sea_level'].mean().reset_index()

    # Plot the monthly averages
    plt.plot(df_monthly_avg['year_month'], df_monthly_avg['sea_level'], label=f'{location} (Monthly Average)', color=color)

    # Plot the yearly average in a contrasting color
    plt.plot(df_yearly_avg['year'], df_yearly_avg['sea_level'], label=f'{location} (Yearly Average)', color=contrast_color, linestyle='--')

    # Labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (mm)')
    plt.title(f'Sea Level in {location} (Monthly & Yearly Average)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Replace -32767 with NaN in all datasets
df2['sea_level'].replace(-32767, np.nan, inplace=True)
df3['sea_level'].replace(-32767, np.nan, inplace=True)
df4['sea_level'].replace(-32767, np.nan, inplace=True)
df5['sea_level'].replace(-32767, np.nan, inplace=True)
df6['sea_level'].replace(-32767, np.nan, inplace=True)
df7['sea_level'].replace(-32767, np.nan, inplace=True)

# Plot each dataset with monthly and yearly averages
plot_with_monthly_and_yearly_average(df2, "Brisbane, Australia", "orange", "black")
plot_with_monthly_and_yearly_average(df3, "Dakar, Senegal", "green", "black")
plot_with_monthly_and_yearly_average(df4, "Hilo, Hawaii", "cyan", "black")
plot_with_monthly_and_yearly_average(df5, "Ushuaia, Argentina", "red", "black")
plot_with_monthly_and_yearly_average(df6, "Nagasaki, Japan", "purple", "black")
plot_with_monthly_and_yearly_average(df7, "Zanzibar, Tanzania", "brown", "black")


```
